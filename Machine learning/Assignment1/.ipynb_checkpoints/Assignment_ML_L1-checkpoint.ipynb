```python
df.hist(figsize=(15,15));
print(df.shape)
```

```python
drop_column=['DailyRate','EmployeeNumber','ID','EmployeeCount','HourlyRate','MonthlyRate','JobLevel']
data=df.drop(drop_column,axis=1)
print(data.shape)

objecttype=[]
for column in data.columns:
  if(df[column].dtype==object):
    objecttype.append(column)
print(objecttype)

print(data.isnull().sum()!=0)
```

```python
for column in objecttype:
  data[column]=data[column].astype('category')
  data[column]=data[column].cat.codes

train = data.drop('Attrition',axis=1)
labels=data['Attrition']
data_corr=train.corr()
plt.figure(figsize=(15,15))
print(train.head())


mask = np.zeros_like(data_corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(data_corr,vmin=-1,vmax=1,cmap='OrRd',mask=mask)
```

```python
from sklearn.preprocessing import StandardScaler
columns=train.columns
scaler = StandardScaler()
train=scaler.fit_transform(train.values)
train=pd.DataFrame(train,columns=columns)
print(train.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report

X_train,X_test,y_train,y_test = train_test_split(train,labels,test_size=0.2,random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
maxi=0
depth=1
for i in range(1,30):
  rfc=RandomForestClassifier(n_estimators=200,max_depth=i,criterion='gini')
  rfc.fit(X_train,y_train)
  y_pred=rfc.predict(X_test)
  if(accuracy_score(y_pred,y_test) > maxi):
    maxi=accuracy_score(y_pred,y_test)
    depth=i
print(depth,maxi)

final_rfc = RandomForestClassifier( n_estimators=200,max_depth=depth, criterion='gini')
final_rfc.fit(X_train,y_train)
y_pred_rfc=final_rfc.predict(X_test)
print(confusion_matrix(y_test,y_pred_rfc))
print(classification_report(y_test,y_pred_rfc))
```

```python
from sklearn.svm import SVC

svc=SVC(kernel='rbf', gamma=0.001,C=85)
svc.fit(X_train,y_train)
y_svc=svc.predict(X_test)
print(accuracy_score(y_svc,y_test))
print(confusion_matrix(y_test,y_svc))
print(classification_report(y_test,y_svc))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train,y_train)
y_pred_KNN=knn.predict(X_test)
print(accuracy_score(y_pred_KNN,y_test))
print(confusion_matrix(y_test,y_pred_KNN))
print(classification_report(y_test,y_pred_KNN))
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pred_LR=lr.predict(X_test)
print(accuracy_score(y_pred_LR,y_test))
print(confusion_matrix(y_test,y_pred_LR))
print(classification_report(y_test,y_pred_LR))
```
